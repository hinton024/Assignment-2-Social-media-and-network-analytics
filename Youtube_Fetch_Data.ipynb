```python
from googleapiclient.discovery 
import build
import pandas as pd
import time
import json

api_key = 'AIzaSyAHcuyAWQG9YVFSlUZwdnKF5G2IJleBWQ0'

# Setting up the YouTube API client
youtube = build('youtube', 'v3', developerKey=api_key)

def fetch_video_details(video_id):
    request = youtube.videos().list(
        part='snippet,statistics,contentDetails',
        id=video_id
    )
    response = request.execute()
    return response['items'][0] if response['items'] else None

# Searching for videos related to 'David Beckham'
request = youtube.search().list(
    q='David Beckham',
    part='snippet',
    type='video',
    maxResults=30  
)
response = request.execute()

# Collecting video details and comments
video_data = []

for item in response['items']:
    video_id = item['id']['videoId']
    video_details = fetch_video_details(video_id)

    if video_details:
        title = video_details['snippet']['title']
        published_at = video_details['snippet']['publishedAt']
        description = video_details['snippet'].get('description', 'N/A')
        category_id = video_details['snippet'].get('categoryId', 'N/A')
        comment_count = video_details['statistics'].get('commentCount', 'N/A')

        # Fetching the top-level comments for the video
        try:
            comments_request = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                maxResults=30
            )
            comments_response = comments_request.execute()

            comments = []
            for comment_item in comments_response.get('items', []):
                comment_text = comment_item['snippet']['topLevelComment']['snippet']['textDisplay']
                comments.append(comment_text)
        except Exception as e:
            comments = []
            print(f"Error fetching comments for video ID {video_id}: {e}")

        # Appending the video information along with comments
        video_data.append({
            "Title": title,
            "Video ID": video_id,
            "Published at": published_at,
            "Description": description,
            "Category ID": category_id,
            "Comment Count": comment_count,
            "Comments": comments
        })

# Printing the video data along with comments
for video in video_data:
    print(f"Title: {video['Title']}")
    print(f"Video ID: {video['Video ID']}")
    print(f"Published at: {video['Published at']}")
    print(f"Description: {video['Description']}")
    print(f"Category ID: {video['Category ID']}")
    print(f"Comment Count: {video['Comment Count']}")
    print("Comments:")
    for comment in video['Comments']:
        print(f"- {comment}")
    print("-" * 125)

# Saving the data to a JSON file
with open('DB_youtube.json', 'w', encoding='utf-8') as json_file:
    json.dump(video_data, json_file, ensure_ascii=False, indent=4)

print("Data saved to 'DB_youtube.json'")
```
